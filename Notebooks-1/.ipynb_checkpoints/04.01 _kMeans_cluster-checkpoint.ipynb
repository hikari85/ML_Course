#### Popular categories of clustering algorithms:
- Partitional clustering. Divides data objects into nonoverlapping groups.
- Density-based clustering. Determines cluster assignments based on the density of data points in a region.  
    Clusters are assigned where there are high densities of data points separated by low-density regions.
- Hierarchical clustering
#### K-Means Clustering
1. Decide the number of **clusters** (k)  
2. Select k random points from the data as **centroids**     
3. Assign all the points to the nearest cluster centroid  
4. Calculate the centroid of newly formed clusters  
5. Repeat steps 3 and 4  until the centroid positions do not change   

Result:
1. The "cluster center" is the arithmetic mean of all the points belonging to the cluster.   
2. Each point is closer to its own cluster center than to other cluster centers.  

```Python
create k random points for starting centroids
while any point has changed cluster assignment
  for every point in the dataset:
    for every centroid:
      calculate distance between the centroid and point
    assign point to the closest cluster
  for every cluster:
    calculate mean of the points in cluster
    assign centroid to mean
```

#### Clustering  
#### Implement the K Means Clustering
In clustering, the data is divided into several groups with similar traits.   
Some clustering algorithms are K-Means, Hierarchical, t-SNE, and DBSCAN   
Try K-Means. 

<u>Finding the best amount of clusters to get most accurate results (KMeans)</u>

**Elbow method** helps to select the optimal number of clusters by fitting the model with a range of values for K in K-means algorithm. Elbow method draws a line plot between SSE (Sum of Squared errors - also termed "inertia") vs number of clusters.   
Affter the _elbow point_ the SSE or inertia starts decreasing in a linear fashion.   
ELBOW RULE - look for the point where the line changes almost 90 degrees.

````python
wcss = [] # Within Cluster Sum of Squares

for k in range(1,7):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(features)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(4,3))
plt.title("WCSS / K Chart")
plt.plot(range(1,7),wcss,"-o")
plt.grid(True)
plt.xlabel("Clusters")
plt.ylabel("Inertia or SSE")
plt.xticks(range(1,7))
plt.tight_layout();
````

````python
# ['%.2f' % elem for elem in wcss]  # previous to f-string
[round(elem, 2)  for elem in wcss]
````

3 or 4 seems to be our best value for clusters. (By the Elbow Rule)  

Clustering adds a new `labels` column to the festures table  

````python
# print(f'before clustering:\n{features.sample(5)}\n')
kmeans = KMeans(n_clusters=3)
features["labels"] = kmeans.fit_predict(features)
print(f'\nlabels added by clustering:\n{features.sample(5)}\n')
````

#### Scatter Plots with value of $k$ from 1 to 4  

````python
# 'sepal_length', 'sepal_width', 'petal_length', 'petal_width'
plt.figure(figsize=(24,4))

plt.suptitle("K Means Clustering",fontsize=20)

plt.subplot(1,5,1)
plt.title("K = 1",fontsize=16)
plt.xlabel("Petal Length Cm")
plt.ylabel("Petal Width Cm")
plt.scatter(features.petal_length,features.petal_width)

# make two clusters
kmeans = KMeans(n_clusters=2)
features["labels"] = kmeans.fit_predict(features)

plt.subplot(1,5,2)
plt.title("K = 2",fontsize=16)
plt.xlabel("Petal Length Cm")
plt.scatter(features.petal_length[features.labels == 0],features.petal_width[features.labels == 0])
plt.scatter(features.petal_length[features.labels == 1],features.petal_width[features.labels == 1])

# make three clusters
kmeans = KMeans(n_clusters=3)
features["labels"] = kmeans.fit_predict(features)

plt.subplot(1,5,3)
plt.title("K = 3",fontsize=16)
plt.xlabel("PetalLengthCm")
plt.scatter(features.petal_length[features.labels == 0],features.petal_width[features.labels == 0])
plt.scatter(features.petal_length[features.labels == 1],features.petal_width[features.labels == 1])
plt.scatter(features.petal_length[features.labels == 2],features.petal_width[features.labels == 2])

# make four clusters
kmeans = KMeans(n_clusters=4)
features["labels"] = kmeans.fit_predict(features)

plt.subplot(1,5,4)
plt.title("K = 4",fontsize=16)
plt.xlabel("PetalLengthCm")
plt.scatter(features.petal_length[features.labels == 0],features.petal_width[features.labels == 0])
plt.scatter(features.petal_length[features.labels == 1],features.petal_width[features.labels == 1])
plt.scatter(features.petal_length[features.labels == 2],features.petal_width[features.labels == 2])
plt.scatter(features.petal_length[features.labels == 3],features.petal_width[features.labels == 3])

# With Original Labels
plt.subplot(1,5,5)
plt.title("Original Labels",fontsize=16)
plt.xlabel("PetalLengthCm")
plt.scatter(df.petal_length[df.species == "Iris-setosa"],df.petal_width[df.species == "Iris-setosa"])
plt.scatter(df.petal_length[df.species == "Iris-versicolor"],df.petal_width[df.species == "Iris-versicolor"])
plt.scatter(df.petal_length[df.species == "Iris-virginica"],df.petal_width[df.species == "Iris-virginica"])

plt.subplots_adjust(top=0.8)
````

#### Cluster the Students  

````python
df = pd.read_excel('../Data/BSc22A_Student_Data.xlsx')
df = df.dropna()
print(f"{df.columns}\nrows:{len(df)}")
````

#### Make Clusters  
Clustering is be based on a combination of features Height, Weight, BMI, and Age.  
Change value of K and repeat clustering   

````python
wcss = [] # Within Cluster Sum of Squares

for k in range(1,8):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(df[['Height_cm', 'Weight_Kg']])
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(4,3))
plt.title("WCSS / K Chart")
plt.plot(range(1,8),wcss,"-o")
plt.grid(True)
plt.xlabel("Clusters")
plt.ylabel("Inertia or SSE")
plt.xticks(range(1,8))
plt.tight_layout()
````

````python
df1 = df[['Name', 'Height_cm', 'Weight_Kg']]
df1['BMI'] = df.Weight_Kg/(df.Height_cm/100)**2
for cluster_count in [3, 4, 5]:    
    kmeans = KMeans(n_clusters=cluster_count)
    df1["label"] = kmeans.fit_predict(df1.loc[:,['Height_cm', 'Weight_Kg', 'BMI']])
    
    clusters = df1.groupby('label')
    print(f"{cluster_count} Clusters")
    for cluster, students in clusters:
        students.BMI = round(students.BMI, 2)
        print(students); print()    
    print(f"{'_'*70}\n")
````

##### You have been clustered mainly by weight

````python
print('\nAggregation of Weight')
df1.groupby('label')['Weight_Kg'].agg(['count', 'min', 'max', 'mean']).sort_values('min')
````

#### 3D Cluster   

````python
from sklearn.datasets import make_blobs # make synthetic dataset
centroids = [(-5,-5,5),(5,5,-5),(3.5,-2.5,4),(-2.5,2.5,-4)]
cluster_std = [1,1,1,1]
X,y = make_blobs(n_samples=200,cluster_std=cluster_std,centers=centroids,n_features=3,random_state=1)
````

````python
# Scatter plot of the dataset
import plotly.express as px
fig = px.scatter_3d(x=X[:,0], y=X[:,1], z=X[:,2])
fig.show()
````

````python
wcss = []
for i in range(1,10):
    km = KMeans(n_clusters=i)
    km.fit_predict(X)
    wcss.append(km.inertia_)

plt.plot(range(1,10),wcss);
````

````python
# Fit the K-Means algorithm with the optimal value of K
km = KMeans(n_clusters=4)
y_pred = km.fit_predict(X)
````

````python
# Analyse the different clusters formed
df = pd.DataFrame()
df['col1'] = X[:,0]
df['col2'] = X[:,1]
df['col3'] = X[:,2]
df['label'] = y_pred

fig = px.scatter_3d(df,x='col1', y='col2', z='col3',color='label')
fig.show()
````
