1. Simple Feature Scaling. Resultant values are in the range of 0 to 1.      
    $$x_i = \frac{x_i}{x_{max}}$$  
E.g., in image data, we know the max value to be 255 and can use the magic number in the denominator.  

2.  Min-Max Scaling gives similar result as simple feature scaling.  
$$x_i = \frac{x_i - x_{min}}{x_{max} - x_{min}}$$   

3. Normalisation changes the observations so that they can be described as a normal distribution. Normalise the data when the model assumes the data to be normally distributed. E.g., linear discriminant analysis (LDA) and Gaussian naive Bayes. Methods with “Gaussian” in the name probably assumes normality.  
$$x_i = \frac{x_i - \mu}{\sigma}$$       

In scaling, the distribution does not change, but the range changes.  
With normalisatin, the distribution changes.  


```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
# from sklearn.preprocessing import normalize

import warnings

warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('../Data/life-expectancy.csv')
df.columns
```

```python
scaler = MinMaxScaler()
normaliser = StandardScaler()

df['life_scaled'] = scaler.fit_transform(df[['Life']])
df['normal_life'] = normaliser.fit_transform(df[['Life']])
# df['normalised_life'] = normalize(df[['Life']])
df['normalised_life'] = (df.Life - df.Life.mean())/df.Life.std()
```

```python
df[['normalised_life','normal_life']]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8, 3))
plt.subplot(2,2,1)
sns.histplot(data = df[['Life']], kde=True)
plt.subplot(2,2,2)
sns.histplot(data = df[['life_scaled']], kde=True)
plt.subplot(2,2,3)
sns.histplot(data = df[['normal_life']], kde=True);
plt.subplot(2,2,4)
sns.histplot(data = df[['normalised_life']], kde=True);
plt.tight_layout();
```
