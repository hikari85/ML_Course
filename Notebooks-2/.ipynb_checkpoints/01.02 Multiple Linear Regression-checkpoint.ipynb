#### Multiple Linear Regression   
Linear Regression approximates a line $\hat{y} = x*w + b$ joining predicted values    
that has the least average error from the actual values.   

If we add a second feature, the example lie in a 3D space.  
$$\hat{y} = x_1*w_1 + x_2*w_2 + b$$
This is the weighted sum of input   


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python

# Import the dataset
x1, x2, x3, y = np.loadtxt("../Data/pizza_3_vars.txt", skiprows=1, unpack=True)

# These weights are from the training phase
w = np.array([-3.98230894, 0.37333539, 1.69202346])

# Plot the axes
sns.set(rc={"axes.facecolor": "white", "figure.facecolor": "white"})
ax = plt.figure().add_subplot(projection="3d") #, figsize=(10, 5))
ax.set_xlabel("Temperature")
ax.set_ylabel("Reservations")
ax.set_zlabel("Pizzas") #, labelpad=5)# , fontsize=30)

# Plot the data points
ax.scatter(x1, x2, y, color='b')

# Plot the plane
MARGIN = 10
edges_x = [np.min(x1) - MARGIN, np.max(x1) + MARGIN]
edges_y = [np.min(x2) - MARGIN, np.max(x2) + MARGIN]
xs, ys = np.meshgrid(edges_x, edges_y)
zs = np.array([w[0] + x * w[1] + y * w[2] for x, y in
               zip(np.ravel(xs), np.ravel(ys))])
ax.plot_surface(xs, ys, zs.reshape((2, 2)), alpha=0.2);
```

##### Loss function without bias   

```python
def loss(X, Y, w):
    return np.average((predict(X, w) - Y) ** 2)
```

##### Upgrade the gradient function  
Upgrade gradient of the loss to multiple variables. 

```python
def gradient(X, Y, w):
    return 2 * np.matmul(X.T, (predict(X, w) - Y))/X.shape[0]
```

```python
def predict(X, w):
    return np.matmul(X, w)

def loss(X, Y, w):
    return np.average((predict(X, w) - Y) ** 2)

def gradient(X, Y, w):
    return 2 * np.matmul(X.T, (predict(X, w) - Y))/X.shape[0]

def train(X,Y,iterations, lr):
    data = []
    w = np.zeros((X.shape[1], 1))
    for i in range(iterations):
        data.append([i, loss(X,Y,w)])
        w -= gradient(X,Y,w)*lr
    return w, pd.DataFrame(data, columns=['i', 'loss'])
```

```python
w, df = train(X,Y,iterations=100000, lr=0.001)
```

```python
df.head()._append(df.tail())
```

#### Add bias   
$$\hat{y} = x_1*w_1 + x_2*w_2 + b$$   
In the implementation so far, we did not add bias.  

Difference between weights and bias: weight is multiplied by X and bias is not multiplied by any value.  
That is, it is multiplied by 1.  

If we rewrite the prediction formula adding $x_0$, with value 1, as the multiplier for bias 
$$\hat{y} = x_1*w_1 + x_2*w_2 + x_0*b$$   
there is no difference between the prediction formula with and without bias.  
Bias becomes a weight of an input variable with value 1.  

 

```python
print(X[:4])
```

```python
X = np.c_[ X, np.ones(X.shape[0]) ]              # add a column
X
```

```python

w_, df_ = train(X,Y,iterations=100000, lr=0.001)
```

```python
print(f"weights without bias:\n{w}\n\nweights with bias:\n{w_}")
```

```python
df_.head()._append(df_.tail())
```

#### Result   
Losses at the end are lower after considering bias.   
The last weight is the bias. 

1. Program extended to handle datasets with multiple input features.    
2. Turned bias into another weight.
