### Decision Tree  

Supervised Model  

A decision tree is an acyclic graph that can be used to make decisions.  
In each branching node of the graph, a specific feature of the feature vector is examined.  
`E.g., feature vecotr: ['age', 'height', 'weight'];  a specific feature: 'weight'  `

If the value of the feature is below a specific threshold, then the left branch is followed;   
otherwise, the right branch is followed.  
`if weight < 60: go_left`   
`else: go_right`

##### Decision Tree  
```
create_branch
  check if every item in the dataset is in the same class
    if so, return the dataset
    else
      find the best feature to split the dataset
      split the dataset
      create a branch node
        for each split
          call create_branch and add the result to the branch node
        return branch node
```                

#### Classification and Regression Trees or CART   
Decision Tree algorithms for classification or regression modeling.

The representation is a binary tree; each node can have zero, one or two child nodes.

A node represents a single input variable (X) and a split point on that variable, assuming the variable is numeric.  
The leaf (terminal) nodes contain an output variable (y) which is used to make a prediction.  

Recursive binary splitting. A numerical procedure where all the values are lined up and different split points are tried and tested using a cost function.

The split with the lowest cost is selected. 

The predicted value $\hat{y}$ for each node is the mode (classification) or mean (regression).

<u>Regression</u>: The cost function that is minimized to choose split points is the sum squared error (RSS) across all training samples.  

<u>Classification</u>: The Gini cost function is used which provides an indication of how pure the nodes are, where node purity refers to how mixed the training data assigned to each node is.  

Splitting continues until nodes contain a minimum number of training examples or a maximum tree depth is reached.

**Ensemble Models** such as   
bagged trees, random forests, and boosted trees,   
are less interpretable but more accurate.

````python
X = iris.data
y = iris.target

# Fit the classifier with default hyper-parameters
clf = DecisionTreeClassifier(random_state=1234)
model = clf.fit(X, y)
````

##### The following code is only for understanding the process/results 

````python
def encode(species):
    if species == 'Iris-setosa':
        species = 0
    elif species == 'Iris-versicolor':
        species = 1
    elif species == 'Iris-virginica':
        species = 2
    return species
````

````python
df1 = df.copy()
df1.species = df1.species.apply(encode)
df1
predicted = model.predict(X)
df1['y_hat'] = predicted
df1['err'] = abs(df1.y_hat - df1.species) 
````

On the training set, decision tree classifier has no prediction error.  

````python
print(df1.head())
print(f"\nCount of wrong classification: {len(df1[df1.err == 1])}")
````

````python
from sklearn.metrics import classification_report
report = classification_report(y, predicted)
print(report)
````

#### Gini coefficient/Gini index/Gini ratio  

Developed by statistician and sociologist Corrado Gini.   
In economics, it is a measure of statistical dispersion representing the income inequality, the wealth inequality, or the consumption inequality within a nation or a social group.

A Gini coefficient of 0 reflects perfect equality, where all income or wealth values are the same, while a Gini coefficient of 1 (or 100%) reflects maximal inequality among values, for example a single individual having all the income while all others have none.

India's [Gini Index](https://www.worldeconomics.com/Inequality/Gini-Coefficient/India.aspx) is 42.9 as measured in 2019. Data Source: World Bank.  

#### 1. Gini Index
Gini index is the cost function to evaluate splits in the dataset.

A split in the dataset involves one input attribute and one value for that attribute  
used to divide training patterns into two groups of rows.   
```
proportion or probability = count(class_value) / count(rows)  
gini_index = 1.0 - sum(proportion * proportion)
```

$$Gini = 1 - \sum_{i=1}^n(p_i)^2$$  

where,  
$p_i$ is the probability of an object being classified to a particular class.

A Gini score gives an idea of how good a split is by how mixed the classes are in the two groups created by the split. A perfect separation results in a Gini score of 0, whereas the worst case split that results in 50/50 classes in each group result in a Gini score of 0.5 (for a 2 class problem).


````python
import numpy as np

def gini(x):
    total = 0
    for i, xi in enumerate(x[:-1], 1):
        total += np.sum(np.abs(xi - x[i:]))
    return total / (len(x)**2 * np.mean(x))
````

````python
incomes1 = np.array([50, 50, 70, 70, 70, 90, 150, 150, 150, 150])
incomes2 = incomes1 * 100

print(gini(incomes1))  
print(gini(incomes2))  
````

#### Decision Trees ALWAYS Overfit   
By default, a decision tree is allowed to grow until all leaves are pure.  
As the model correctly classifies ALL training instances, this leads to:  
- 100% overfitting, and  
- poor generalization  

Cost-complexity-pruning \(CCP\) is an effective technique to prevent this.  
CCP considers a combination of two factors for pruning a decision tree:  
- Cost \(C\): Number of misclassifications  
- Complexity \(C\): Number of nodes  
The core idea is to iteratively drop sub-trees, which, after removal, lead to:  
- a minimal increase in classification cost  
- a maximum reduction of complexity (or nodes)  
If two sub-trees lead to a similar increase in classification cost, then it is wise to remove the sub-tree with more nodes.  
In sklearn, we can control cost-complexity-pruning using the ccp_alpha parameter:  
- large value of ccp_alpha → results in underfitting  
- small value of ccp_alpha → results in overfitting  
The objective is to determine the optimal value of ccp_alpha, which gives a better model.  

````python
tree = DecisionTreeClassifier(ccp_alpha=0).fit(X,y)
# check training and test scores
#  decistion tree regression plot?
````
